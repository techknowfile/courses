```python
import math
import numpy as np
import matplotlib.pyplot as plt
%pylab inline


def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))
# evenly sampled time at 200ms intervals
x = np.arange(-1, 4, 0.1)
gauss = 2/3*1/(math.sqrt(2*math.pi))*math.e**(-1*(x-1)**2/2)
y = np.arange(0, 3, 0.1)
myline = [1/9 for x in y]
fig = plt.figure()
plt.plot(x, gauss)
plt.plot(y, myline, color='orange')
plt.plot([2.3212, 2.3212], [0, 1/9], '-', color='orange')
plt.axvline(0, color='black')
plt.axhline(0, color='black')
plt.plot(2.3212, 1/9, 'ro')
plt.plot(3, 0.035993977675458701300376133607142387826543026964572204978, 'ro')
fig.savefig('./images/plot.png')
plt.show()
```
